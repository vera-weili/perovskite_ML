```python
# data_train = pd.read_csv('train_data.csv')
y = data_clean['vbm_diff'].values
X = data_clean.drop(["vbm_diff"], axis=1).values
features = data_clean.drop(["vbm_diff" ], axis=1).columns
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold

reg = GradientBoostingRegressor(max_depth=10, n_estimators=20)
skf = KFold(n_splits=3, shuffle=True, random_state=1)
for train, test in skf.split(X, y):
    reg.fit(X[train], y[train])
    Y_test_pred = reg.predict(X[test])
    mse = mean_squared_error(y[test], Y_test_pred)
    mae = mean_absolute_error(y[test], Y_test_pred)
    #test_auc = auc(y[test], Y_test_pred)
    #test_accb = balanced_accuracy_score(y[test], Y_test_pred)
    print("mean_squared_error", mse, "mean_absolute_error", mae)#, "Auc", test_auc)#, "Balanced Accuracy", test_accb)
    print("Feature Importance:", features[np.argsort(reg.feature_importances_)][::-1])
    print("Feature Importance:", reg.feature_importances_)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
reg = GaussianProcessRegressor(random_state=0)
skf = KFold(n_splits=3, shuffle=True, random_state=1)
for train, test in skf.split(X, y):
    scaler.fit(X[train])
    reg.fit(scaler.transform(X[train]), y[train])
    Y_test_pred = reg.predict(scaler.transform(X[test]))
    mse = mean_squared_error(y[test], Y_test_pred)
    mae = mean_absolute_error(y[test], Y_test_pred)
    #test_auc = auc(y[test], Y_test_pred)
    #test_accb = balanced_accuracy_score(y[test], Y_test_pred)
    print("mean_squared_error", mse, "mean_absolute_error", mae)#, "Auc", test_auc)#, "Balanced Accuracy", test_accb)
```

```python
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
reg = KernelRidge(alpha=0, kernel="sigmoid")
skf = KFold(n_splits=3, shuffle=True, random_state=1)
for train, test in skf.split(X, y):
    scaler.fit(X[train])
    reg.fit(scaler.transform(X[train]), y[train])
    Y_test_pred = reg.predict(scaler.transform(X[test]))
    mse = mean_squared_error(y[test], Y_test_pred)
    mae = mean_absolute_error(y[test], Y_test_pred)
    #test_auc = auc(y[test], Y_test_pred)
    #test_accb = balanced_accuracy_score(y[test], Y_test_pred)
    print("mean_squared_error", mse, "mean_absolute_error", mae)#, "Auc", test_auc)#, "Balanced Accuracy", test_accb)
```
